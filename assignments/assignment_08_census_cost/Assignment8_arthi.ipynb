- Cleansing the data to make the year, population, cost, cost per person numerical type
- Code functions to cleanse the columns. One function for one column
- Using apply() method of data frame to apply the functions to the columns for data cleansing

```python
# Function to cleanse the year

def cleanse_year(a_string):
    return int(a_string.replace("*",""))



df["YEAR"] = df["Census Year"].apply(cleanse_year)

df
```

```python
# Function to convert a str of numbers separated by commas into a integer

def cleanse_population(pop_string):
    return int(pop_string.replace(",",""))


df["POPULATION"] = df["Total Population"].apply(cleanse_population)

df

df.info()
```

```python
# Function to convert the cost which is seperated by commas to a numeric type
def cleanse_total_cost(cost_string):

  cleansed_cost = cost_string.replace('$', '')
  cleansed_cost = cleansed_cost.replace(',', '')

  if "Billion" in cleansed_cost:
      cleansed_cost = cleansed_cost.replace(" Billion", "")
      final_cost = float(cleansed_cost) * 1000000000
  else:
      final_cost = float(cleansed_cost)

  return final_cost
  



df["Census Cost"] = df["Census Cost"].apply(cleanse_total_cost)
df


```

```python
def cleanse_avg_cost(cost_string):
    
    if "cents" in cost_string:
        cost = float(cost_string.replace(" cents", ""))
    else:
        cost = float(cost_string.replace("$", "")) * 100
    
    return cost


df["COST PER PERSON"] = df["Average Cost Per Person"].apply(cleanse_avg_cost)
df
```

- Using Plotly, plotting the average cost per person over time.

```python
import plotly.express as px  
```

```python
fig = px.bar(
    df, 
    x = "YEAR", 
    y = "COST PER PERSON", 
    title = "Average Cost Per Person over Time", 
    color = "YEAR", 
    template = "plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```

- Using Plotly, plotting the scatter plot of Total Population vs Total Cost.

```python
fig = px.scatter(
    df,
    x="POPULATION",
    y="Census Cost", 
    title="Total Population vs Total Cost", 
    color="Census Cost", 
    template="presentation"
)

fig.update_layout(showlegend=True)
fig.update_layout(
    xaxis_title="Total Population",
    yaxis_title="Total Cost",)


fig.show()
```
