```python
import numpy as np
import pandas as pd

pd.__version__
pd.set_option("display.max_rows", 200)   # to be able to see all rows
```

```python
data_url = "https://raw.githubusercontent.com/sruthib22/data690_1/main/doubt_file/bowling_stats.md"

df = pd.read_csv(data_url, header = None)

df.head()
```

```python
df.columns = ['year', 'state','state_code','count1','count2']

df
```

```python
# Use Numpy's isnan() function to test if a number is a NaN (Not a Number)

def add(count1, count2):

    if np.isnan(count1):    
        return np.nan
    elif np.isnan(count2):
        return count1
    else:
        return count1 * 1000 + count2


# Combining a lambda function with a user-defined add() function. 
# This allows for operations on multiple columns
# Quite tricky!

df["count"] = df[["count1", "count2"]].apply(lambda x: add(x.count1, x.count2),axis=1) 

df
```
