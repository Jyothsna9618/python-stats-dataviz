```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(2022)
X_train = rng.normal(0, 10, 70)
print(X_train)
```

```python
plt.figure(figsize=(10,6))
plt.hist(X_train)
plt.show()
```

```python
y_train = 10 * X_train ** 2 + rng.normal(0,100,70)
print(y_train)
```

```python
plt.figure(figsize=(10,6))
plt.hist(y_train)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=X_train, y=y_train)
plt.show()
```
